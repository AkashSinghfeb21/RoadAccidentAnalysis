# Encode categorical variables if needed.Handle outliers and duplicate entries.


```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Columns to encode
label_cols = [
   'Country', 'State', 'City', 'County',
    'Timezone', 'Airport_Code',
    'Weather_Condition', 'Wind_Direction',
    'Sunrise_Sunset', 'Civil_Twilight', 
    'Nautical_Twilight', 'Astronomical_Twilight',
    'Weekday'
]

# Decide a threshold for low vs high cardinality
threshold = 20  # if unique values <= 20 → one-hot, else label encode

# Loop through categorical columns
for col in label_cols:
    if col not in df.columns:
        continue
    if df[col].nunique() <= threshold:
        # One-hot encode (safe for small categories)
        dummies = pd.get_dummies(df[col], prefix=col, drop_first=True)
        df = pd.concat([df, dummies], axis=1)
        df.drop(col, axis=1, inplace=True)
    else:
        # Label encode (better for large categories)
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col].astype(str))

#one hot encode which create dummies columns or extra columns 
#of unique values from original column ie country column is given then 
#one hot encode will create dummies columns as countryfrance countryusa countryindia
#and each values are converted into binary that is either 0 or 1. 0 for value notexists
#1 for values exists in original column and each represented separately in each columns that is
#countryusa countryfrance countryindia

# #drop first = true means if the values are 0 on both side then it will 1 on first column that is france in this ex case 
# #which will remove redundant values or column

# Display the first few rows of the encoded DataFrame
df.head()
```

```python
# 1. Identify numeric columns for outlier detection

# We don’t check categorical columns like Country, Road Type, Accident Cause. Focus on continuous numeric features:
# Numerical features (continuous, usable for outlier detection, scaling, etc.)

# [Distance(mi),Temperature(F),Wind_Chill(F),Humidity(%),Pressure(in)

# Visibility(mi),Wind_Speed(mph),Precipitation(in), Hour,Month,Year, Latitude,Longitude]

import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(y=df['Visibility(mi)'])
plt.title('Boxplot of Visibility (with outliers)')
plt.show()
```

```python
# # Count duplicates
# print("Duplicate rows:", df.duplicated().sum())

# # Drop duplicates
# df = df.drop_duplicates()

#there is no duplicate so this line didnot needed

def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)   # 25% value
    Q3 = df[column].quantile(0.75)   # 75% value
    IQR = Q3 - Q1                    # range in middle 50%
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    # Keep only values within range
    cleaned_df = df[(df[column] >= lower) & (df[column] <= upper)]
    
    # Plot cleaned column
    sns.boxplot(y=cleaned_df[column])
    plt.title(f'Boxplot of {column} (without outliers)')
    plt.show()
    
    return cleaned_df


#df.info()
```

# Accident Hotspots alert system

Step 1: Identify Hotspots from Your Dataset

```python
from sklearn.cluster import DBSCAN
import numpy as np

# Load dataset (assuming df is your DataFrame)
sample_df = df[['Latitude', 'Longitude']].dropna().sample(n=50000, random_state=42)
coords = sample_df.to_numpy()

# coords = df[['Latitude', 'Longitude']].dropna().to_numpy()

# DBSCAN for hotspot detection
kms_per_radian = 6371.0088  # Earth radius in km
epsilon = 1.5 / kms_per_radian  # 1.5 km radius
db = DBSCAN(eps=epsilon, min_samples=20, algorithm='ball_tree', metric='haversine').fit(np.radians(coords))

sample_df['cluster'] = db.labels_

# Get hotspot centers
hotspots = sample_df.groupby('cluster')[['Latitude', 'Longitude']].mean().reset_index()
hotspots = hotspots[hotspots['cluster'] != -1]  # remove noise
```

```python
import winsound
import tkinter as tk
from tkinter import messagebox
def alert_driver(hotspot, dist):
    """
    Trigger alarm + popup when near hotspot
    """
    # 🔊 Play sound alert (Windows)
    try:
        winsound.Beep(1000, 1000)  # frequency=1000Hz, duration=1s
    except:
        print("Sound not supported on this OS")

    # 💬 Popup message
    root = tk.Tk()
    root.withdraw()  # hide main window
    messagebox.showwarning("⚠️ Accident Hotspot Alert",
                           f"You are {dist:.2f} km from a known accident hotspot!\n"
                           f"Hotspot Location: {hotspot}")
    root.destroy()
```

```python
from geopy.distance import geodesic

def is_near_hotspot(driver_lat, driver_lon, hotspots, threshold=1.0):
    for _, row in hotspots.iterrows():
        hotspot = (row['Latitude'], row['Longitude'])
        dist = geodesic((driver_lat, driver_lon), hotspot).km
        if dist <= threshold:  # within 500 meters
            alert_driver(hotspot,dist)
            return True, hotspot
    return False, None
```

```python
import folium
from folium import CircleMarker


# Start map centered at mean coordinates
m = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=6)

# Add hotspots
for _, row in hotspots.iterrows():
    folium.CircleMarker(location=[row['Latitude'], row['Longitude']],
                        radius=8, color="red", fill=True, fill_opacity=0.6,
                        popup="Accident Hotspot").add_to(m)

m.save("hotspots_map.html")

#open hotspots_map.html to see hotspots in browser
```
