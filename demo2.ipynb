# Encode categorical variables if needed.Handle outliers and duplicate entries.


```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Columns to encode
label_cols = [
   'Country', 'State', 'City', 'County',
    'Timezone', 'Airport_Code',
    'Weather_Condition', 'Wind_Direction',
    'Sunrise_Sunset', 'Civil_Twilight', 
    'Nautical_Twilight', 'Astronomical_Twilight',
    'Weekday'
]

# Decide a threshold for low vs high cardinality
threshold = 20  # if unique values <= 20 → one-hot, else label encode

# Loop through categorical columns
for col in label_cols:
    if col not in df.columns:
        continue
    if df[col].nunique() <= threshold:
        # One-hot encode (safe for small categories)
        dummies = pd.get_dummies(df[col], prefix=col, drop_first=True)
        df = pd.concat([df, dummies], axis=1)
        df.drop(col, axis=1, inplace=True)
    else:
        # Label encode (better for large categories)
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col].astype(str))

#one hot encode which create dummies columns or extra columns 
#of unique values from original column ie country column is given then 
#one hot encode will create dummies columns as countryfrance countryusa countryindia
#and each values are converted into binary that is either 0 or 1. 0 for value notexists
#1 for values exists in original column and each represented separately in each columns that is
#countryusa countryfrance countryindia

# #drop first = true means if the values are 0 on both side then it will 1 on first column that is france in this ex case 
# #which will remove redundant values or column

# Display the first few rows of the encoded DataFrame
df.head()
```

```python
# 1. Identify numeric columns for outlier detection

# We don’t check categorical columns like Country, Road Type, Accident Cause. Focus on continuous numeric features:
# Numerical features (continuous, usable for outlier detection, scaling, etc.)

# [Distance(mi),Temperature(F),Wind_Chill(F),Humidity(%),Pressure(in)

# Visibility(mi),Wind_Speed(mph),Precipitation(in), Hour,Month,Year, Latitude,Longitude]

import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(y=df['Visibility(mi)'])
plt.title('Boxplot of Visibility (with outliers)')
plt.show()
```

```python
# # Count duplicates
# print("Duplicate rows:", df.duplicated().sum())

# # Drop duplicates
# df = df.drop_duplicates()

#there is no duplicate so this line didnot needed

def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)   # 25% value
    Q3 = df[column].quantile(0.75)   # 75% value
    IQR = Q3 - Q1                    # range in middle 50%
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    # Keep only values within range
    cleaned_df = df[(df[column] >= lower) & (df[column] <= upper)]
    
    # Plot cleaned column
    sns.boxplot(y=cleaned_df[column])
    plt.title(f'Boxplot of {column} (without outliers)')
    plt.show()
    
    return cleaned_df


#df.info()
```

```python
df_no_outliers = remove_outliers(df, 'Visibility(mi)')

print("Before:", df.shape)
print("After:", df_no_outliers.shape)
```

```python

# Load the columns you need
df = pd.read_csv(
    "USroadaccident/Us_Accidents.csv",
    usecols=["Severity", "Weather_Condition"]
)

# Make sure Weather_Condition is treated as a string
df["Weather_Condition"] = df["Weather_Condition"].astype(str)

# Keep rows mentioning Rain or Fog
mask = df["Weather_Condition"].str.contains("Rain|Fog", case=False, na=False)
rain_fog = df.loc[mask].copy()

# Simplify all variations into 2 groups
def simplify_weather(w):
    return "Rain" if "Rain" in w else "Fog"

rain_fog["WeatherGroup"] = rain_fog["Weather_Condition"].apply(simplify_weather)

# Bar plot of mean Severity
plt.figure(figsize=(6, 5))
sns.barplot(
    data=rain_fog,
    x="WeatherGroup",
    y="Severity",
    palette="Set2",
    ci="sd"          # show SD as error bar
)
plt.title("Average Accident Severity during Rain vs Fog")
plt.xlabel("Weather")
plt.ylabel("Average Severity")
plt.ylim(0, rain_fog["Severity"].max() + 0.5)
plt.show()
```
