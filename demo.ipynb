```python
# First, ensure Month names are converted to proper numbers
df['Month'] = pd.to_datetime(df['Month'], format='%B').dt.month  # e.g., "October" → 10

# Create a base date using Year + Month
df['Date'] = pd.to_datetime(df[['Year', 'Month']].assign(Day=1))

# # Map Day of Week to numbers (Monday=0, Sunday=6)
# day_map = {
#     "Monday": 0, "Tuesday": 1, "Wednesday": 2, "Thursday": 3,
#     "Friday": 4, "Saturday": 5, "Sunday": 6
# }
# df['Day of Week Num'] = df['Day of Week'].map(day_map)

# Map Time of Day to rough hour slots (customize if needed)
time_map = {
    "Morning": 9,       # around 9 AM
    "Afternoon": 15,    # around 3 PM
    "Evening": 19,      # around 7 PM
    "Night": 23         # around 11 PM
}
df['Hour'] = df['Time of Day'].map(time_map)

# Now build a proper datetime
df['Datetime'] = pd.to_datetime(
    df[['Year', 'Month']].assign(Day=1)
) + pd.to_timedelta(df['Hour'], unit='h')

# Drop helper columns if not needed
# df.drop(columns=['Day of Week Num', 'Hour'], inplace=True)
```

# Encode categorical variables if needed.Handle outliers and duplicate entries.


```python
import pandas as pd

#one hot encode which create dummies columns or extra columns 
#of unique values from original column ie country column is given then 
#one hot encode will create dummies columns as countryfrance countryusa countryindia
#and each values are converted into binary that is either 0 or 1. 0 for value notexists
#1 for values exists in original column and each represented separately in each columns that is
#countryusa countryfrance countryindia

# List of categorical columns to be one-hot encoded
label_cols = [
    'Country', 'Region', 'Weekday', 'Time of Day',
    'Urban/Rural', 'Road Type', 'Weather Conditions',
    'Road Condition', 'Accident Cause'
]

# One-Hot Encode categorical variables
df_encoded = pd.get_dummies(df, columns=label_cols, drop_first=True)

#drop first = true means if the values are 0 on both side then it will 1 on first column that is france in this ex case 
#which will remove redundant values or column

# Display the first few rows of the encoded DataFrame
df_encoded.head()

#hotspot accident area feature proposed by sri shanya.
```

```python
#2 handling outliers
import numpy as np

num_cols = ['Visibility Level', 'Traffic Volume', 'Medical Cost', 'Economic Loss']

for col in num_cols:
    z_scores = (df[col] - df[col].mean()) / df[col].std()
    df = df[np.abs(z_scores) < 3]  # Keep only values within 3 std
```

```python
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower) & (df[col] <= upper)]
```
