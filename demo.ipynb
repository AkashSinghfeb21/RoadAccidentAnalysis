```python
# 1. Identify numeric columns for outlier detection

# We don’t check categorical columns like Country, Road Type, Accident Cause. Focus on continuous numeric features:

#  ["Visibility Level", "Speed Limit", "Driver Alcohol Level",
#                 "Traffic Volume", "Medical Cost", "Economic Loss", "Population Density"]

import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(y=df['Speed Limit'])
plt.title('Boxplot of Speed Limit (with outliers)')
plt.show()
```

```python
# # Count duplicates
# print("Duplicate rows:", df.duplicated().sum())

# # Drop duplicates
# df = df.drop_duplicates()

#there is no duplicate so this line didnot needed

def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)   # 25% value
    Q3 = df[column].quantile(0.75)   # 75% value
    IQR = Q3 - Q1                    # range in middle 50%
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    # Keep only values within range
    cleaned_df = df[(df[column] >= lower) & (df[column] <= upper)]
    
    # Plot cleaned column
    sns.boxplot(y=cleaned_df[column])
    plt.title(f'Boxplot of {column} (without outliers)')
    plt.show()
    
    return cleaned_df


#df.info()
```

```python
df_no_outliers = remove_outliers(df, 'Speed Limit')

print("Before:", df.shape)
print("After:", df_no_outliers.shape)
```

```python
numeric_cols = ["Visibility Level", "Speed Limit", "Driver Alcohol Level",
                "Traffic Volume", "Medical Cost", "Economic Loss", "Population Density"]

df_clean = df.copy()
for col in numeric_cols:
    df_clean = remove_outliers(df_clean, col)

print("Final shape after cleaning:", df_clean.shape)
```

```python
#Analyze distribution of accident severity levels.

import matplotlib.pyplot as plt
import seaborn as sns

# Count plot for accident severity
plt.figure(figsize=(6,4))
sns.countplot(x='Accident Severity', data=df, palette="Set2")
plt.title("Distribution of Accident Severity Levels")
plt.xlabel("Accident Severity")
plt.ylabel("Count")
plt.show()

# Pie chart
severity_counts = df['Accident Severity'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(severity_counts, labels=severity_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set2"))
plt.title("Accident Severity Distribution")
plt.show()
```

```python
#Plot accident frequency by time of day, day of the week, and month.

# Time of Day
plt.figure(figsize=(6,4))
sns.countplot(x='Time of Day', data=df, order=df['Time of Day'].value_counts().index, palette="muted")
plt.title("Accidents by Time of Day")
plt.ylabel("Count")
plt.show()

# Day of Week
plt.figure(figsize=(6,4))
sns.countplot(x='Weekday', data=df, palette="coolwarm")
plt.title("Accidents by Weekday (0=Mon ... 6=Sun)")
plt.ylabel("Count")
plt.show()

# Month
plt.figure(figsize=(8,4))
sns.countplot(x='Month', data=df, palette="crest")
plt.title("Accidents by Month")
plt.ylabel("Count")
plt.show()
```

```python
# Examine most common weather conditions and road types during accidents.  

# Weather Conditions
plt.figure(figsize=(8,4))
sns.countplot(y='Weather Conditions', data=df, order=df['Weather Conditions'].value_counts().index, palette="viridis")
plt.title("Accidents by Weather Conditions")
plt.xlabel("Count")
plt.ylabel("Weather Condition")
plt.show()

# Road Types
plt.figure(figsize=(8,4))
sns.countplot(y='Road Type', data=df, order=df['Road Type'].value_counts().index, palette="plasma")
plt.title("Accidents by Road Type")
plt.xlabel("Count")
plt.ylabel("Road Type")
plt.show()
```

```python
# Use histograms, pie charts, and bar graphs. 

numeric_cols = ['Visibility Level', 'Speed Limit', 'Emergency Response Time',
                'Traffic Volume', 'Medical Cost', 'Economic Loss']

df[numeric_cols].hist(bins=30, figsize=(12,8), color="skyblue", edgecolor="black")
plt.suptitle("Histograms of Numeric Features", size=16)
plt.show()
```

# WEEK 4 Bivariate and Mulvariate Analysis

```python
#Analyze correlation between severity and:- 

# Map Accident Severity into numbers
severity_map = {'Low': 1, 'Medium': 2, 'High': 3}
df['Severity_Num'] = df['Accident Severity'].map(severity_map)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select numeric columns we care about
corr_cols = ['Severity_Num', 'Visibility Level', 'Traffic Volume']
corr = df[corr_cols].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Weather conditions- Visibility- Road surface condition- Trafic congestion 

# Visibility vs Severity
plt.figure(figsize=(6,4))
sns.boxplot(x='Accident Severity', y='Visibility Level', data=df)
plt.title("Accident Severity vs Visibility Level")
plt.show()

# Traffic Volume vs Severity
plt.figure(figsize=(6,4))
sns.boxplot(x='Accident Severity', y='Traffic Volume', data=df)
plt.title("Accident Severity vs Traffic Volume")
plt.show()


```

```python
# Weather Conditions
plt.figure(figsize=(8,5))
sns.countplot(x='Accident Severity', hue='Weather Conditions', data=df)
plt.title("Accident Severity vs Weather Conditions")
plt.show()

# Road Condition
plt.figure(figsize=(8,5))
sns.countplot(x='Accident Severity', hue='Road Condition', data=df)
plt.title("Accident Severity vs Road Condition")
plt.show()
```
